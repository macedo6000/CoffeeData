```python
# Add Matplotlib inline magic command
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load 
data_to_load = "loan_data.csv"


# Covert to Dataframe
loan_data_df = pd.read_csv(data_to_load)
```

```python
loan_data_df.info()
```

```python
agency_df = loan_data_df[['agency_code','agency_name', 'agency_abbr']]

agency_df.drop_duplicates(inplace=True, ignore_index=True)
agency_df.sort_values(by=['agency_code'], inplace=True, ignore_index=True)
agency_df.set_index('agency_code', inplace=True)

agency_count = loan_data_df.groupby('agency_code')['agency_code'].count()
agency_df['count']= agency_count

agency_df.to_csv('agency.csv')
agency_df
```

```python
loan_type_df = loan_data_df[['loan_type_name', 'loan_type']]
loan_type_df.drop_duplicates(inplace=True, ignore_index=True)
loan_type_df.sort_values(by=['loan_type'], inplace=True)
loan_type_df.set_index('loan_type', inplace=True)

loan_type_count = loan_data_df.groupby('loan_type')['loan_type'].count()
loan_type_df['count'] = loan_type_count

loan_type_df.to_csv('loan_type.csv')
loan_type_df
```

```python
property_df = loan_data_df[['property_type_name', 'property_type']]
property_df.drop_duplicates(inplace=True, ignore_index=True)
property_df.sort_values(by=['property_type'], inplace=True)
property_df.set_index('property_type', inplace=True)

property_count = loan_data_df.groupby('property_type')['property_type'].count()
property_df['count'] = property_count

property_df.to_csv('property.csv')
property_df
```

```python
purpose_df = loan_data_df[['loan_purpose_name', 'loan_purpose']]
purpose_df.drop_duplicates(inplace=True, ignore_index=True)
purpose_df.sort_values(by=['loan_purpose'], inplace=True,ignore_index=True)
purpose_df.set_index('loan_purpose', inplace=True)

purpose_count = loan_data_df.groupby('loan_purpose')['loan_purpose'].count()
purpose_df['count'] = purpose_count

purpose_df.to_csv('purpose.csv')
purpose_df
```

```python
owner_occupancy_df = loan_data_df[['owner_occupancy_name', 'owner_occupancy']]
owner_occupancy_df.drop_duplicates(inplace=True, ignore_index=True)
owner_occupancy_df.sort_values(by=['owner_occupancy'], inplace=True,ignore_index=True)
owner_occupancy_df.set_index('owner_occupancy', inplace=True)

owner_occupancy_count = loan_data_df.groupby('owner_occupancy')['owner_occupancy'].count()
owner_occupancy_df['count'] = owner_occupancy_count

owner_occupancy_df.to_csv('owner_occupancy.csv')
owner_occupancy_df
```
